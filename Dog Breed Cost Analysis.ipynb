# Merge

```python
def split_ailments(ailments):
    return [ailment.strip() for ailment in ailments.split(',')]
```

```python
genetic = pd.read_csv('gen_of_dog.csv')
price_gen = pd.read_excel('price_of_gen.xlsx')

price_gen['GENETIC_AILMENTS'] = price_gen['GENETIC_AILMENTS'].str.lower().str.strip()
genetic['GENETIC_AILMENTS'] = genetic['GENETIC_AILMENTS'].str.lower().str.strip()

#แยกชื่อโรคแต่ละแถว
genetic['AILMENT_LIST'] = genetic['GENETIC_AILMENTS'].apply(split_ailments)

#แยกชื่อออกเป็นแถวตัวเอง
genetic_ex = genetic.explode('AILMENT_LIST')

genetic_ex.rename(columns={'AILMENT_LIST': 'GEN_TOTAL'}, inplace=True)

dog_m = pd.merge(genetic_ex, price_gen, left_on='GEN_TOTAL', right_on='GENETIC_AILMENTS', how='inner')

dog_m.head()
```

```python
dog_m.to_csv('dog_m.csv', index=False)
```
